# Horizontal Bar Chart – Top 5 Products by Sales: Displays the five highest-grossing products, highlighting category leaders in total sales amount.

1. Tablets lead sales performance with the highest revenue (~31.5K), suggesting strong market demand. This category could benefit from expanded promotions or inventory to sustain momentum.

2. Wearables (Smartwatches) are close behind (~28.2K), indicating a growing consumer shift toward tech-lifestyle products. Bundling with smartphones or accessories could drive cross-sell opportunities.

3. Laptops (~27.7K) show steady sales, highlighting them as a consistent revenue driver. Targeted offers for students and professionals might further boost this category.

4. Apparel (T-Shirts & Jackets) contribute significantly (~27K each), proving fashion is an important parallel revenue stream. Seasonal collections or discounts could amplify sales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Set 'date' as index and resample by month, summing sales_amount
monthly_sales = df.set_index('date').resample('M')['sales_amount'].sum()

# Create enhanced line chart
plt.figure(figsize=(12, 6))

# Plot with enhanced styling
line = plt.plot(monthly_sales.index, monthly_sales.values,
                color='#2E8B57',  # Sea green color
                linewidth=3,      # Thick line
                marker='o',       # Circle markers
                markersize=8,     # Larger markers
                markerfacecolor='#90EE90',  # Light green markers
                markeredgecolor='#2E8B57',  # Dark green marker edges
                markeredgewidth=2,
                alpha=0.9)

# Add data point value labels (every other point to avoid clutter)
for i, (date, value) in enumerate(monthly_sales.items()):
    if i % 2 == 0:  # Show every other label
        plt.annotate(f'${value:,.0f}',
                    (date, value),
                    textcoords="offset points",
                    xytext=(0,15),
                    ha='center',
                    fontsize=10,
                    fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.3',
                             facecolor='white',
                             edgecolor='#2E8B57',
                             alpha=0.8))

# Enhance chart appearance
plt.xlabel('Month', fontsize=14, fontweight='bold')
plt.ylabel('Sales Amount ($)', fontsize=14, fontweight='bold')
plt.title('Monthly Sales Trends', fontsize=16, fontweight='bold', pad=20)

# Format x-axis dates
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))

# Rotate x-axis labels 45 degrees
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Format y-axis to show currency
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x:,.0f}'))

# Enhanced grid
plt.grid(True, alpha=0.3, linestyle='-', linewidth=0.8, color='gray')
plt.grid(True, which='minor', alpha=0.1, linestyle=':', linewidth=0.5)

# Add trend line (optional - shows overall trend)
if len(monthly_sales) > 2:
    z = np.polyfit(range(len(monthly_sales)), monthly_sales.values, 1)
    p = np.poly1d(z)
    plt.plot(monthly_sales.index, p(range(len(monthly_sales))),
             "--", alpha=0.6, color='red', linewidth=2, label='Trend Line')
    plt.legend(loc='upper left', fontsize=12)

# Optimize for dark/light themes
plt.rcParams.update({
    'axes.facecolor': 'white' if plt.rcParams['figure.facecolor'] == 'white' else 'black',
    'axes.edgecolor': 'black' if plt.rcParams['figure.facecolor'] == 'white' else 'white',
    'text.color': 'black' if plt.rcParams['figure.facecolor'] == 'white' else 'white',
    'xtick.color': 'black' if plt.rcParams['figure.facecolor'] == 'white' else 'white',
    'ytick.color': 'black' if plt.rcParams['figure.facecolor'] == 'white' else 'white'
})

# Add subtle background shading for better visual appeal
ax.set_facecolor('#f8f9fa' if plt.rcParams['figure.facecolor'] == 'white' else '#2d2d2d')

# Improve layout and spacing
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)  # Extra space for rotated labels
plt.show()

# Display enhanced monthly sales summary
print("Monthly Sales Summary:")
monthly_sales_df = monthly_sales.reset_index()
monthly_sales_df.columns = ['Month', 'Total Sales']
monthly_sales_df['Total Sales'] = monthly_sales_df['Total Sales'].apply(lambda x: f'${x:,.0f}')
monthly_sales_df['Month'] = monthly_sales_df['Month'].dt.strftime('%B %Y')
print(monthly_sales_df.to_string(index=False))

# Print summary statistics
print(f"\nSummary Statistics:")
print(f"Average Monthly Sales: ${monthly_sales.mean():,.0f}")
print(f"Highest Month: {monthly_sales.idxmax().strftime('%B %Y')} (${monthly_sales.max():,.0f})")
print(f"Lowest Month: {monthly_sales.idxmin().strftime('%B %Y')} (${monthly_sales.min():,.0f})")
print(f"Total Sales Period: ${monthly_sales.sum():,.0f}")
plt.savefig('trend.png')   # <-- saves as PNG file
plt.show()
# Save the current plot
plt.savefig('Monthly Sales Trends', dpi=300, bbox_inches='tight')
plt.savefig('Monthly Sales Trends', dpi=300, bbox_inches='tight')  # Also save as PDF
```

# Line Chart – Monthly Sales Trends: Shows overall revenue fluctuations across months, revealing seasonality and growth patterns.

August 2024 recorded the peak sales at $27,860, about 29% above the monthly average—this suggests strong seasonal demand; replicating August’s promotions or campaigns could help replicate this spike in other months.

June 2024 was the weakest month at $16,672, nearly 24% below average, signaling a mid-year slump; targeted campaigns or discount strategies in Q2 could help stabilize sales during this dip.

Q3 (July–September) was the strongest quarter ($75,332), consistently outperforming the average; this period could be leveraged for major product launches or bundled offers.

Sales volatility is evident (range of $11,188 between lowest and highest month), highlighting the need for demand forecasting and inventory planning to manage stock better.

With total sales of $262,991 across the year, maintaining the average monthly run rate (~$22K) while addressing weak months could push annual revenue beyond $280K next year.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Group by region and calculate mean customer_age
avg_age_by_region = df.groupby('region')['customer_age'].mean()

# Create upgraded vertical bar chart
plt.figure(figsize=(8, 6))

# Create orange-to-red gradient colors
n_bars = len(avg_age_by_region)
colors = plt.cm.OrRd(np.linspace(0.5, 0.9, n_bars))

# Create bars with enhanced styling
bars = plt.bar(avg_age_by_region.index, avg_age_by_region.values,
               color=colors,
               edgecolor='white',
               linewidth=1.5,
               alpha=0.9)

# Add value labels on top of bars with enhanced styling
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + max(avg_age_by_region) * 0.01,
             f'{height:.1f}',
             ha='center', va='bottom',
             fontsize=12,
             fontweight='bold',
             bbox=dict(boxstyle='round,pad=0.3',
                      facecolor='white',
                      edgecolor=colors[i],
                      alpha=0.8,
                      linewidth=1))

# Enhanced chart styling
plt.xlabel('Region', fontsize=12, fontweight='bold')
plt.ylabel('Average Customer Age (Years)', fontsize=12, fontweight='bold')
plt.title('Average Customer Age by Region', fontsize=14, fontweight='bold', pad=20)

# Improve tick styling
plt.xticks(rotation=45, fontsize=12, ha='right')
plt.yticks(fontsize=12)

# Add subtle grid for better readability
plt.grid(axis='y', alpha=0.3, linestyle='--', linewidth=0.8)

# Optimize for dark/light theme compatibility
ax = plt.gca()

# Dynamic theme detection and styling
is_light_theme = plt.rcParams['figure.facecolor'] == 'white'

if is_light_theme:
    ax.set_facecolor('#f8f9fa')
    text_color = 'black'
    grid_color = 'gray'
else:
    ax.set_facecolor('#2d2d2d')
    text_color = 'white'
    grid_color = 'lightgray'

# Apply theme colors
plt.rcParams.update({
    'text.color': text_color,
    'axes.labelcolor': text_color,
    'xtick.color': text_color,
    'ytick.color': text_color
})

# Add age range annotations for context
min_age = avg_age_by_region.min()
max_age = avg_age_by_region.max()
age_range = max_age - min_age

# Add a subtle horizontal line showing overall average
overall_avg = avg_age_by_region.mean()
plt.axhline(y=overall_avg, color='red', linestyle=':', alpha=0.7, linewidth=2,
           label=f'Overall Average: {overall_avg:.1f} years')

# Add legend for the average line
plt.legend(loc='upper right', fontsize=11, framealpha=0.9)

# Enhance y-axis formatting
y_max = max(avg_age_by_region) * 1.15
plt.ylim(0, y_max)

# Add subtle bar shadows for depth
for i, bar in enumerate(bars):
    shadow_bar = plt.bar(bar.get_x() + 0.02, bar.get_height() - 0.5,
                        bar.get_width(), 0.5,
                        color='black', alpha=0.1, zorder=0)

# Improve layout with better spacing
plt.tight_layout()
plt.subplots_adjust(bottom=0.15, top=0.9)  # Space for rotated labels and title
plt.show()

# Enhanced data display with statistics
print("Average Customer Age by Region:")
print("=" * 40)

# Create a formatted summary table
age_summary = avg_age_by_region.reset_index()
age_summary.columns = ['Region', 'Avg Age']
age_summary['Avg Age'] = age_summary['Avg Age'].round(1)
age_summary = age_summary.sort_values('Avg Age', ascending=False)

for idx, row in age_summary.iterrows():
    region = row['Region']
    age = row['Avg Age']
    print(f"{region:.<20} {age:>6.1f} years")

print("=" * 40)
print(f"Overall Statistics:")
print(f"  Highest: {age_summary.iloc[0]['Region']} ({age_summary.iloc[0]['Avg Age']:.1f} years)")
print(f"  Lowest:  {age_summary.iloc[-1]['Region']} ({age_summary.iloc[-1]['Avg Age']:.1f} years)")
print(f"  Range:   {age_range:.1f} years")
print(f"  Average: {overall_avg:.1f} years")
plt.show()
# Save the current plot
plt.savefig('Average Customer Age by Region', dpi=300, bbox_inches='tight')
plt.savefig('Average Customer Age by Region', dpi=300, bbox_inches='tight')  # Also save as PDF
```
